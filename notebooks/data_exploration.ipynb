```python
import os

import polars as pl

pl.Config.set_fmt_str_lengths(100)
pl.Config.set_fmt_float("full")
```

```python
# clinical notes
clinical_df = pl.read_csv("/gpfs/data/razavianlab/longchen/brain_mri/data/nyu_mri_dataset/full/clinical_notes_batch012_20250605.csv")
```

```python
# radiology notes
mri_radiology_notes = "/gpfs/data/razavianlab/longchen/brain_mri/data/nyu_mri_dataset/full/radiology_notes_batch012_20250605.csv"
ct_radiology_notes = "/gpfs/data/razavianlab/capstone/2025_stroke/ct_accession_report_2506_deid.csv"

radiology_df = pl.read_csv(
    ct_radiology_notes,
    infer_schema_length=10000,
)
# .with_columns(
#     accession_num=pl.col("accession_num").cast(pl.Int64),
# )
```

```python
datapath = "/gpfs/scratch/gs4342/ct_datasets_backup/nyu"

files = {
    "test": "nyu_test_label.csv",
    "train": "nyu_train_label.csv",
    "val": "nyu_val_label.csv",
}

train_fp = os.path.join(datapath, files["train"])
test_fp = os.path.join(datapath, files["test"])
val_fp = os.path.join(datapath, files["val"])

train_df = pl.read_csv(train_fp)
test_df = pl.read_csv(test_fp)
val_df = pl.read_csv(val_fp)

image_df = pl.concat([train_df, test_df, val_df], how="vertical").with_columns(
    pl.col("accession_num").cast(pl.String)
)
```
